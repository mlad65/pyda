## Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
movies = pd.read_csv(r'ml-25m/movies.csv')
ratings = pd.read_csv(r'ml-25m/ratings.csv')
```

```python
ratings.head()
```

```python
movies.head()
```

```python
movies = movies.merge(ratings, 'inner', on='movieId', suffixes=('', ''))
```

```python
movies.query('rating == 5').groupby(['movieId', 'title']).count().sort_values(by='rating', ascending=False).reset_index()[['movieId', 'title', 'rating']]
```

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
countries = power['country'].unique()
countries.sort()
countries
```

```python
(power.query('(category == 4 | category == 12 | category == 21) &' + 
            '(year >= 2005 & year <= 2010 & quantity > 0) & ' +
            '(country == "Estonia" | country == "Lithuania" | country == "Latvia")')
 ['quantity'].sum()
)
```
